### 4Sum

```python
def fourSum(A, target):
    n = len(A)
    ans = set()
    A.sort()

    for i in range(n):
        if i > 0 and A[i] == A[i-1]:
            continue
        
        for j in range(i+1, n):
            if j > i+1 and A[j] == A[j-1]:
                continue
            k = j+1
            l = n-1

            while k<l:
                sum = A[i] + A[j] + A[k] + A[l]
                if sum < target:
                    k += 1
                elif sum > target:
                    l -= 1
                else:
                    ans.add((A[i], A[j], A[k], A[l]))
                    k += 1
                    l -= 1

                    while (k<l and A[k] == A[k-1]):
                        k += 1
                    while (k<l and A[l] == A[l+1]):
                        l -= 1
    return list(ans)

nums = [1,0,-1,0,-2,2]
target = 0

fourSum(nums, target)
```

```python
nums = [
    15, -3, 7, 20, 5, 9, -11, 4, 8, 2, -6, 13, -7, 1, -2, 18,
    14, 6, -9, 3, 12, -5, 10, -1, 17, 16, 0, -4, -8, 11, -10, 19,
    # Injected quadruplets for target = 42:
    5, 10, 15, 12,     # 5 + 10 + 15 + 12 = 42
    6, 7, 8, 21,       # 6 + 7 + 8 + 21 = 42
    9, 11, 13, 9       # 9 + 11 + 13 + 9 = 42
]
target = 42
fourSum(nums, target)
```
