### Insert In Stack

```python
#User function Template for python3

class Solution:
    def InsertInStack(self,n,arr):
        #code here
        stack = []
        for i in arr:
            stack.append(i)
        return stack


#{ 
 # Driver Code Starts
#Initial Template for Python 3

#contributed by RavinderSinghPB
if __name__=='__main__':
    tcs=int(input())

    for _ in range(tcs):
        n=int(input())
        arr=[int(x) for x in input().split()]

        stack=Solution().InsertInStack(n,arr)

        while stack:
            print(stack.pop(),end=' ')
        print()

        print("~")
# } Driver Code Ends
```

### Implement Stack Using Array

```python
#{ 
 # Driver Code Starts
#Initial Template for Python 3



stackMax=100000
stack=[-1]*stackMax
top=-1



# } Driver Code Ends

#Back-end complete function Template for Python 3

def push(data):
    global stackMax
    global top
    global stack
    top+=1

    if(top>=stackMax):
        top-=1
        print("Stack Full")
    else:
        stack[top]=data
      

def pop():
    global stack
    global stackMax
    global top
    if(top<=-1):
        print("Stack Empty")
    else:
        stack[top]=-1
        top-=1


        
        
def display():
    global stack
    global stackMax
    global top
    if(top<=-1):
        print(-1)
    else:
        i=0
        limit=top if top<stackMax else stackMax-1
        while(i<=top):
            print(stack[i],end=" ")
            i+=1
        print()





#{ 
 # Driver Code Starts.



def main():
    global top
    testcases=int(input())
    while(testcases>0):
        top=-1
        queries=int(input())
        while(queries>0):
            
            queryType=input()
            
            if(queryType[0]=='1'):
                
                x=int(queryType[2])
                push(x)
                
            elif(queryType[0]=='2'):
                pop()
            
            else:
                display()
            
            queries-=1
        
        testcases-=1

        print("~")
if __name__=="__main__":
    main()
# } Driver Code Ends
```
